## LETS PREDICT BASELINE REVENUE ESTIMATE 

We can calculate a baseline estimate of the electricity revenue for every customer for the next twelve
months based on the forecast consumption and forecast price and actual churn outcome. Call this
basecase_revenue

```python
test_df["churn"].value_counts(normalize=True)
```

```python
# Electricity revenue for each customer consists of energy consumption (amount* price) and the meter rent
# Note that we need to reverse the log10-transformation from the data cleaning␣step that we did to remove ositive skewness
#power function in numpy just simpy make power of given element in dataframe 
test_df['basecase_revenue'] = (np.power(10, test_df['forecast_cons_12m']) + 1)* test_df['forecast_price_energy_off_peak'] +test_df['forecast_meter_rent_12m']
#Taking churn or people left into consideration
test_df['basecase_revenue_after_churn'] = test_df['basecase_revenue'] * (1 -  test_df['churn'])
```
